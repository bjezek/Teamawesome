```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

# reading data
df = pd.read_csv("../archive/TweetsElonMusk.csv")
df_lines = df.groupby('name').count()
df_lines = df_lines.sort_values(by=['tweet'], ascending=False)[:10]
names = df_lines.index.values

# filtering out non-top characters
df_sentiment = df[df['name'].isin(names)]
df_sentiment = df_sentiment[['name', 'tweet']]

# calculating sentiment score
sid = SentimentIntensityAnalyzer()
df_sentiment.reset_index(inplace=True, drop=True)
df_sentiment[['neg', 'neu', 'pos', 'compound']] = df_sentiment['tweet'].apply(sid.polarity_scores).apply(pd.Series)
df_sentiment
```
